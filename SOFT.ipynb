```python
import numpy as np
import matplotlib.pyplot as plt
def softmax(x):
  '''Compute softmax values for each sets of scores in x.'''
  return np.exp(x) / np.sum(np.exp(x), axis=0)
x=np.linspace(-10,10)
plt.plot(x, softmax(x))
plt.axis('tight')
plt.title("Actiation function :Softmax")
plt.show()
```
