```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
def sigmoid(x):
                return 1/(1+np.exp(-x))
x=np.linspace(-10,10)
plt.plot(x,sigmoid(x))
plt.axis('tight')
plt.title('Activation function :sigmoid')
plt.show()
```
