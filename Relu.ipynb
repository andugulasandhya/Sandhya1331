```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
def RELU(x):
  x1=[]
  for i in x:
    if i<0:
      x1.append(0)
    else:
       x1.append(i)
  return x1
x=np.linspace(-10,10)
plt.plot(x,RELU(x))
plt.axis('tight')
plt.show()
```
